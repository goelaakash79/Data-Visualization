```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('local_weather_data.csv')
# data.head(1000)
```

```python
# year18 = data[data.Year == 2018]
# year18.head()
```

```python
year_sorted = sorted(set(data.Year))
```

```python
for given_year in year_sorted:
    year = data[data.Year == given_year]
    
    plt.title(given_year)
    plt.scatter(year.OutHum, year.TempOut, 1)
    plt.xlabel('Humidity')
    plt.ylabel('Temprature')
    plt.savefig(str(given_year), dpi=200)
    plt.show()
```
