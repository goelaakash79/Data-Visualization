```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('countries.csv')
```

```python
us = data[data.country == 'United States']
```

```python
china = data[data.country == 'China']
```

```python
us_population = us.population
china_population = china.population
```

```python
us_population_growth = us_population / us_population.iloc[0] * 100
china_population_growth = china_population / china_population.iloc[0] * 100
```

```python
plt.title("Population Growth in US and China (first year = 100)")
plt.plot(us.year, us_population_growth)
plt.plot(china.year, china_population_growth)
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('Population Growth')
plt.show()
```
