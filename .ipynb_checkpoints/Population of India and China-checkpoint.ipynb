```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('countries.csv')
data.head()
```

```python
india = data[data.country == 'India']
china =  data[data.country == 'China']
```

```python
plt.plot(india.year, india.population)
plt.show()
```

```python
india_growth = india.population / india.population.iloc[0] * 100
china_growth = china.population / china.population.iloc[0] * 100
```

```python
plt.title("India and China Population Comparsion")
plt.plot(india.year, india_growth)
plt.plot(china.year, china_growth)
plt.xlabel("year")
plt.ylabel("Population Growth")
plt.legend(['India', 'China'])
plt.show()
```
